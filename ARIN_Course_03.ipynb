```python
# Data from Website
file_url = 'http://winterolympicsmedals.com/medals.csv'
medaldata = pd.read_csv(file_url)
```

```python
# Visualization the first ten values
medaldata.head(10)
```

**Data Types Uniqueness**

```python
print('Countries: ')
medaldata['NOC'].unique()
```

```python
print('Sport types: ')
medaldata['Sport'].unique()
```

```python
print('Discipline types: ')
medaldata['Discipline'].unique()
```

**Golden Medals of Finnland**

```python
# Extracting data of Finnland
print('Golden Medals of Finnland')
medaldata[(medaldata['NOC'] == 'FIN') & (medaldata['Medal'] == 'Gold')]
```

**Women' Golden Medals of Finnland on Icehockey**

```python
# Extracting data of Finnland
print("Women' Golden Medals of Finnland")
medaldata[(medaldata['NOC'] == 'FIN') & (medaldata['Medal'] == 'Gold') & (medaldata['Sport'] == 'Skiing') & (medaldata['Event gender'] == 'W')]
```

**Women' Silver Medals of Finnaland on Figure Skating**

```python
# Extracting data of Finnland
print("Women' Silver Medals of Finnland")
medaldata[(medaldata['NOC'] == 'FIN') & (medaldata['Medal'] == 'Silver') &  (medaldata['Event gender'] == 'W') & (medaldata['Discipline'] == 'Figure skating')]
```
